# Data Visualizations

```python
sns.set(style="whitegrid")
```

# Univariate Analysis

```python
# 1. Distribution of Price
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

**Insight**: This plot helps visualize the distribution of prices. You can determine if prices are skewed or normally distributed.

```python
# 2. Distribution of Rating
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=10, kde=True, color="green")
plt.title("Distribution of Rating")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

**Insight**: This reveals the spread of ratings, which can indicate how customers feel about the products.

```python
# Rating distribution with violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x=df['Rating'], color='green')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.show()
```

**Insight**: Violin plot shows the density of the ratings and provides more information about distribution shape. It's particularly useful for showing multimodal distributions.

```python
# 3. Distribution of Discount
plt.figure(figsize=(10, 6))
sns.histplot(df['Discount'], bins=10, kde=True, color="orange")
plt.title("Distribution of Discount")
plt.xlabel("Discount (%)")
plt.ylabel("Frequency")
plt.show()
```

**Insight**: You can observe the discount spread and how many products are being offered with heavy discounts.

```python
# Discount distribution with bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=df['Discount'].value_counts().index, y=df['Discount'].value_counts().values, color='orange')
plt.title('Discount Distribution')
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.show()
```

**Insight:** Bar plot provides a visual of how many products fall under certain discount categories.

```python
# Battery distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Battery'], kde=True, color='red')
plt.title('Battery Capacity Distribution')
plt.xlabel('Battery Capacity (mAh)')
plt.ylabel('Frequency')
plt.show()
```

**Insight**: It helps understand the typical battery capacity of the phones in the dataset.

# Bivariate Analysis

```python
# Price vs Rating with scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Price', y='Rating', hue='Brand', palette='viridis')
plt.title('Price vs Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

**Insight**: Scatter plot helps visualize if higher-priced phones correlate with better ratings. We can also see how different brands perform.

```python
# Price vs Discount
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Price', y='Discount', hue='Brand', palette='coolwarm')
plt.title('Price vs Discount')
plt.xlabel('Price')
plt.ylabel('Discount (%)')
plt.show()
```

**Insight**: Helps assess if higher-priced phones are getting larger discounts.

# Multivariate Analysis

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
corr_matrix = df[['Price', 'Rating', 'Discount', 'Battery']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Price, Rating, Discount, and Battery')
plt.show()
```

# Categorical Analysis

```python
# Brand distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Brand', data=df, palette='Set2')
plt.title('Brand Distribution')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**Insight**: This plot shows how the different brands are distributed in the dataset.

```python
# Pairplot with 'Brand' as hue
sns.pairplot(df, hue="Brand", vars=["Price", "Rating", "Discount"], palette="Set2")
plt.suptitle('Pairplot of Price, Rating, and Discount by Brand', y=1.02)
plt.show()
```

```python
# Create Rating Categories
df['Rating Category'] = pd.cut(df['Rating'], bins=[0, 2.5, 3.5, 4.5, 5], labels=['Low', 'Average', 'High', 'Very High'])

# Heatmap of Brand vs Rating Categories and Discount
pivot_table = pd.pivot_table(df, values="Discount", index="Brand", columns="Rating Category", aggfunc="mean")
sns.heatmap(pivot_table, annot=True, cmap="coolwarm")
plt.title('Heatmap of Brand vs Rating Categories and Discount')
plt.show()
```
