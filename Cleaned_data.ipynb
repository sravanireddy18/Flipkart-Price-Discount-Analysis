```python
import pandas as pd #To load and manipulate the dataset
import numpy as np
import matplotlib.pyplot as plt #For creating visualizations
import seaborn as sns # For statistical plots
```

```python
df = pd.read_csv(r"C:\Users\bsrav\INNOMATICS\INNOMATICS_NOTEBOOK\PROJECT\Flipkart_webscraping_prj\flipkart_products_1501.csv")
df
```

```python
# Rename 'Processor' to 'Display' and 'Display' to 'Processor'
df.rename(columns={"Processor": "Temp", "Display": "Processor"}, inplace=True)
df.rename(columns={"Temp": "Display"}, inplace=True)
```

```python
df.head()
```

```python
# Remove unwanted characters from Price and convert to numeric
df["Price"] = df["Price"].str.replace("₹", "").str.replace(",", "").str.strip()
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
```

```python
#Clean Discount: Remove '%' and convert to float
df["Discount"] = df["Discount"].str.replace("% off", "").str.strip()
df["Discount"] = pd.to_numeric(df["Discount"], errors="coerce")
```

```python
# Clean Rating: Remove '★' symbol and convert to float
df["Rating"] = df["Rating"].str.replace("★", "").str.strip()
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
```

```python
# Handle missing values
df.fillna("Unknowm", inplace=True)

# Strip extra spaces from text columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Step 1: Split 'Product Name' into 'Brand' and 'Model'
df['Brand'] = df['Product Name'].str.extract(r'([A-Za-z]+)')[0]
df['Model'] = df['Product Name'].str.replace(r'^[A-Za-z]+\s', '', regex=True)
```

```python
df
```

```python
df['Battery'] = df['Battery'].str.extract(r'(\d+)').astype(float)
```

```python
df
```

```python
# Step 4: Convert the columns to correct datatypes
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')
df['Battery'] = pd.to_numeric(df['Battery'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
```

```python
# Filling NaN values in Discount with 0
df['Discount'] = df['Discount'].fillna(0)
```

```python
# Save cleaned data to a new CSV file
df.to_csv("flipkart_products_cleaned1.csv", index=False)

# Print cleaned data summary
print("Data cleaning complete!")
print(df.info())  # Check data types
print(df.head())  # Preview first 5 rows
```
